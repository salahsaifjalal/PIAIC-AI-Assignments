```python
from tkinter import *

my_root=Tk()

my_root.configure(bg="green")

tasks=[]
tasks=["Fajr Prayer", "Zuhr Prayer", "Asr Prayer", "Maghrib Prayer", "Isha Prayer"]

def update_listbox():
    clear_listbox()
    for x in tasks:
        my_List_Box.insert("end",x)
        
def clear_listbox():
    my_List_Box.delete(0,"end")

def show_task():
    update_listbox()

def add_task():
    task=my_Input.get()
    if task!="":
        tasks.append(task)
        update_listbox()
    else:
        my_Label2["text"]="PLEASE ENTER A TASK."
    my_Input.delete(0,"end")
        
        
def delete_task():
    my_task=my_List_Box.get("active")
    if my_task in tasks:
        tasks.remove(my_task)
    update_listbox()
    

def delete_all_task():
    global tasks
    tasks=[]
    update_listbox()



my_root.minsize(500,600)
my_root.maxsize(500,600)

my_root.title('To Do List Program Made by Salahuddin Muhammad')

my_Label1=Label(my_root, text="To-Do-List-Program", font="courgette 16 bold", bg="green", fg="white")
my_Label1.pack(pady=5)

my_Label2=Label(my_root, text="", bg="green")
my_Label2.pack(pady=5)

my_Input=Entry(my_root, width=70, borderwidth=5, bg="white", fg="red")
my_Input.pack(pady=5)

Show_Task_button=Button(my_root, text="Press me to view task.", font="elephant 11 bold", fg="blue", bg="white", command=show_task)
Show_Task_button.pack(pady=5)


Add_Task_button=Button(my_root, text="Press me to add task.",font="elephant 11 bold", fg="green", bg="white", command=add_task)
Add_Task_button.pack(pady=5)

Delete_Task_button=Button(my_root, text="Press me to delete selected task.",font="elephant 11 bold", fg="purple", bg="white", command=delete_task)
Delete_Task_button.pack(pady=5)

Delete_All_Task_button=Button(my_root, text="Press me to delete all task.",font="elephant 11 bold", fg="orange", bg="white", command=delete_all_task)
Delete_All_Task_button.pack(pady=5)

Exit_Task_button=Button(my_root, text="Press me to Exit.",font="elephant 11 bold", fg="red", bg="white", command=my_root.destroy)
Exit_Task_button.pack(pady=5)

my_List_Box=Listbox(my_root, height=15, width=70)
my_List_Box.pack(pady=5)

my_root.mainloop()
```
